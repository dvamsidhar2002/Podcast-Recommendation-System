```python
#importing the dependecies
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.feature_extraction.text import TfidfVectorizer
from sklearn.metrics.pairwise import cosine_similarity
```

```python
#loading the dataset into the workspace
data = pd.read_csv('poddf.csv')
data.head()
```

```python
podcast = data[['Name','Genre','Description']]
podcast.head()
```

```python
podcast.shape
```

```python
combine_features = data['Name'] + ' ' + data['Genre'] + ' ' + data['Description']
```

```python
print(combine_features)
```

```python
vectorizer = TfidfVectorizer()
feature_vectors = vectorizer.fit_transform(combine_features)
```

```python
similarity = cosine_similarity(feature_vectors)
```

```python
print(similarity)
```

```python
import difflib
podcast_Name = str(input('Enter the name of your favorite podcast : '))
list_of_all_names = data['Name'].tolist()
find_close_match = difflib.get_close_matches(podcast_Name, list_of_all_names)
print(find_close_match)
```

```python
close_match = find_close_match[0]
print(close_match)
```

```python
index_of_the_podcast_name = data[data['Name'] == close_match]['index'].values[0]
print(index_of_the_podcast_name)
```

```python
similarity_score = list(enumerate(similarity[index_of_the_podcast_name]))
```

```python
sorted_similar_podcast_name = sorted(similarity_score, key = lambda x:x[1], reverse = True) 
print(sorted_similar_podcast_name)
```

```python
print('Podcast recommended for you : \n')

i = 1

for name in sorted_similar_podcast_name:
  index = name[0]
  title_from_index = data[data.index==index]['Name'].values[0]
  if (i<10):
    print(i, '.',title_from_index)
    i+=1
```
